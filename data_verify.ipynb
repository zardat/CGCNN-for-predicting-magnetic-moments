```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file without headers
df = pd.read_csv("my_new_data/id_prop_All.csv", header=None)
print(df.shape)
# Plot the distribution of the third column (index 2)
plt.figure(figsize=(6, 4), dpi=300)
plt.hist(df.iloc[:, 2], bins=30, color='skyblue', edgecolor='black')
plt.title("DS-1a Magnetic Moment Distribution", fontsize=15)
plt.xlabel("Magnetic Moment", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tick_params(axis='both', labelsize=14)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Load the CSV file
data = pd.read_csv('directory/id_prop_Mo_FM.csv', header=None)

# Display the original value range for column C (index 2)
print("Original range in column C:")
print("Min:", data[2].min())
print("Max:", data[2].max())

# Apply clipping at 100, then log transformation, then scaling by 2
data[2] = np.log10(data[2].clip(upper=100)) / 2

# Display the new value range after clipping, log, and scaling
print("\nRange in column C after clipping, log, and scaling:")
print("Min:", data[2].min())
print("Max:", data[2].max())

# Save the processed data to a new CSV file
data.to_csv('directory/cliplog_Mo_FM2.csv', index=False, header=False)
```

```python
import pandas as pd
import numpy as np

# Load the CSV file
data = pd.read_csv('directory/id_prop_Mn_FM.csv', header=None)

# Display the original value range for column C (index 2)
print("Original range in column C:")
print("Min:", data[2].min())
print("Max:", data[2].max())
print(data[2].head())

# Apply the log transformation
data[2] = np.log10(data[2])

# Display some initial values after log transformation
print("\nSome initial values after log transformation:")
print("Min:", data[2].min())
print("Max:", data[2].max())
print(data[2].head())

# Normalize by dividing by the maximum value after log transformation
max_value = data[2].max()
data[2] = data[2] / max_value

# Display the final value range after normalization
print("\nRange in column C after normalization:")
print("Min:", data[2].min())
print("Max:", data[2].max())

# Display some final normalized values
print("\nSome final normalized values:")
print(data[2].head())

# Save the processed data to a new CSV file
data.to_csv('directory/logscaled_Mn_FM.csv', index=False, header=False)
```

```python
import pandas as pd
import numpy as np

# Load the CSV file
data = pd.read_csv('directory/id_prop_Mn_FM.csv', header=None)

# Display the original value range for column C (index 2)
print("Original range in column C:")
print("Min:", data[2].min())
print("Max:", data[2].max())

# Filter out rows where column C values are greater than 100
data = data[data[2] <= 100]

# Apply log transformation (no clipping, just log)
data[2] = np.log10(data[2])

# Display the new value range after log transformation
print("\nRange in column C after filtering and log transformation:")
print("Min:", data[2].min())
print("Max:", data[2].max())

# Save the processed data to a new CSV file
data.to_csv('directory/cliplg_Mn_FM1.csv', index=False, header=False)
```
